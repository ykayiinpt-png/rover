```python
import numpy as np
import scipy as sp

import matplotlib.pyplot as plt
```

```python
velo_filename = "inputs/velocity.txt"
```

```python
with open(velo_filename, "r") as f:
    data = f.readlines()
```

```python
data = [float(v.strip()) for v in data]
```

```python
data[59:70]
```

```python
data = np.array(data)   # your sampled data

dt = 0.1                 # sampling interval (seconds)
fs = 1 / dt              # = 10 Hz
N = len(data)

# Remove DC offset
data = data - np.mean(data)

# Apply window (recommended)
window = np.hanning(N)
data = data * window

# FFT
fft_vals = np.fft.fft(data)
freqs = np.fft.fftfreq(N, d=dt)

# Keep positive frequencies
half = N // 2
freqs = freqs[:half]
magnitude = np.abs(fft_vals[:half]) * 2 / N

# Plot
plt.plot(freqs, magnitude)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.title("FFT Spectrum")
plt.grid()
plt.show()
```
